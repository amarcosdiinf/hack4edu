If the sentiment of the post is negative, but the number of likes is sueperior to the mean likes per post, then it will be considered as neuter.

### Creation of emojis arrays in order to map the text when there are emojis in it

```python
positive_emojis = ["", "", "", "わ", "А", "", "", "", "", "", "", "", "ワ", "猴", "", "", "", ""]
negative_emojis = [""]
```

## Get comments

```python
if os.path.isfile("corpus.csv"): #We already have the corpus
    df = pd.read_csv("corpus.csv")
    comments_dict = df.stack().groupby(level=1).agg(list).to_dict()
else:                            #We don't have the corpus, so we scrap the comments from Instagram (~ 4 hours)
    n_requests = 0
    comments_dict = {}

    medias = instagram.get_medias("profuturo_", account.media_count)

    for media in medias:
        comments_text = []

        comments = instagram.get_media_comments_by_id(media.identifier, 30)
        for comment in comments["comments"]:
            comments_text.append(comment.text)

        comments_dict[media.identifier] = comments_text

        if n_requests < 10:
            n_requests += 1
            sleep(30)
        else:
            n_requests = 0
            sleep(300)
```

## Saving the corpus

```python
df = pd.DataFrame({key: pd.Series(value) for key, value in comments_dict.items()})
df.to_csv("corpus.csv", encoding='utf-8', index=False)
```

# Calculate sentiment

```python
sentiment = sentiment_analysis.SentimentAnalysisSpanish()

sentiment_dict = {}

for media_id, comments_list in comments_dict.items():
    sentiment_result = "neuter"
    n_comments = len(comments_list)
    #print(f"\nPublicaci贸n con id {media_id}:")
    #print(f"{n_comments} comments:")
    sentiment_grade = 0
    mean_sentiment_grade = 0
    for comment in comments_list:
        # If the comments contain emojis, it is assumed that the text will express the same sentiment as the emojis do.
        if any(emoji in comment for emoji in positive_emojis):
            comment = "perfecto"
        elif any(emoji in comment for emoji in negative_emojis):
            comment = "mal"
        sentiment_grade += sentiment.sentiment(comment)
        
    if n_comments > 0:
        mean_sentiment_grade = sentiment_grade/n_comments
    else:
        mean_sentiment_grade = -1
        
    #print(f"Sentimiento medio de la publicaci贸n: {mean_sentiment_grade}")
    
    if (mean_sentiment_grade > 0.7) or ((mean_sentiment_grade == -1) and (media.likes_count > mean_likes_per_post)):
        #print("El sentimiento de la publicaci贸n es positivo.")
        sentiment_result = "positive"
    elif mean_sentiment_grade == -1:
        #print("La publicaci贸n no tiene comentarios, por lo tanto no tiene sentimiento.")
        sentiment_result = "none"
    #elif (media.likes_count > mean_likes_per_post) or (mean_sentiment_grade >= 0.2 and mean_sentiment_grade <= 0.7):
        #print("El sentimiento de la publicaci贸n es neutro.")
    #    sentiment_result = "neuter"
    elif (mean_sentiment_grade >= 0.2 and mean_sentiment_grade <= 0.7):
        #print("El sentimiento de la publicaci贸n es neutro.")
        sentiment_result = "neuter"
    else:
        #print("El sentimiento de la publicaci贸n es negativo.")
        sentiment_result = "negative"
        
    sentiment_dict[media_id] = sentiment_result
    #print('--------------------------------------------------')

#print("\nSentimientos de las distintas publicaciones de ProFuturo: ")
#for media_id, sentiment in sentiment_dict.items():
#    print(f"Id publicaci贸n: {media_id} - Sentimiento: {sentiment}")
```

## Progression number of uploaded posts

```python
media_id_list = []
date_list = []
for media in medias:
    media_id_list.append(media.identifier)
    date_list.append(pd.Timestamp(media.created_time, unit='s').date())
                     
upload_date_dict = dict({"media_id": media_id_list, "upload_date": date_list})
upload_date_df = pd.DataFrame(data=upload_date_dict)
```

```python
upload_date_df["upload_year_month"] = pd.to_datetime(upload_date_df["upload_date"]).dt.to_period('M')
```

```python
upload_date_grouped_df = upload_date_df.groupby(["upload_year_month"]).size().sort_values(ascending=False).reset_index(name="count")
upload_date_grouped_df.astype({"upload_year_month": "string"}).dtypes
```

```python
upload_date_grouped_df.sort_values(by=["upload_year_month"], ignore_index=True, inplace=True)
```

```python
upload_date_grouped_df
```

## Word Cloud

### Obtaining the text

```python
comments_list = []
for comment_value in comments_dict.values():
    if len(comment_value) > 0:
        for comment in comment_value:
            comments_list.append(comment)
```

```python
text = " ".join(comment for comment in comments_list)
```

```python
print (f"There are {len(text)} words in the combination of all review.")
```

### Creating the word cloud

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["la", "el", "de", "con", "por", "del", "que", "y", "e", "se", "en", "es", "lo", "un", 
                  "las", "para", "pero", "mi", "su", "los", "son", "mis", "nos", "una", "esta", "m谩s", 
                  "ser", "nuestro", "mas", "esa", "as铆", "tan", "Qu茅", "al", "este", "le", "est谩", "esto",
                 "ello", "como", "sin", "nuestra", "profuturo", "profuturo_"])
```

```python
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
```

```python
# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,6), dpi=80)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
